# **PRACTICA DE CLASE**



```python

# Se define la clase Perpceptron
class Perpceptron:
    # Constructor: se ejecuta al crear un objeto de la clase
    def __init__(self, pesos, sesgo, tasa_de_aprendizaje):
        # Inicializamos los pesos, el sesgo (bias) y la tasa de aprendizaje
        self.pesos = pesos
        self.sesgo = sesgo
        self.tasa_de_aprendizaje = tasa_de_aprendizaje

    # FUNCIÓN DE ACTIVACIÓN
    # Convierte el valor 'z' en una salida binaria

    def funcion_de_activacion(self, z):
        # Si la suma ponderada es mayor o igual que 0 → salida = 1
        if z >= 0:
            return 1
        # Si es menor que 0 → salida = 0
        else:
            return 0

    # ENTRENAMIENTO DEL PERCEPTRÓN

    def entrenamiento(self, x, y, epocas):
        # Se repite el proceso de aprendizaje varias veces (epocas)
        for epoca in range(epocas):
            print(f"\n=== Época {epoca + 1} ===")

            # Se recorren todos los datos de entrada (x)
            for i, dato in enumerate(x):
                # Se obtienen las dos entradas del vector de datos
                x1, x2 = dato
                # Se extraen los pesos actuales
                w1, w2 = self.pesos
                # Se toma el sesgo actual
                b = self.sesgo

                # Se calcula la combinación lineal:
                #    z = (x1*w1) + (x2*w2) + b

                z = x1 * w1 + x2 * w2 + b

                #  Se aplica la función de activación
                y_pred = self.funcion_de_activacion(z)

                #  Se obtiene la salida real esperada
                y_real = y[i]


                #  Se actualizan los pesos y el sesgo:
                #    w = w + tasa * (y_real - y_pred) * x
                #    b = b + tasa * (y_real - y_pred)

                self.pesos[0] = w1 + self.tasa_de_aprendizaje * (y_real - y_pred) * x1
                self.pesos[1] = w2 + self.tasa_de_aprendizaje * (y_real - y_pred) * x2
                self.sesgo = b + self.tasa_de_aprendizaje * (y_real - y_pred)

                # Se imprimen los valores actualizados
                print(f"Dato: {dato}, Predicción: {y_pred}, Pesos: {self.pesos}, Sesgo: {self.sesgo}")


#     DATOS DE ENTRENAMIENTO (compuerta AND)


# Entradas: todas las combinaciones posibles de 2 bits
x = [[1,1], [1,0], [0,1], [0,0]]
# Salidas esperadas de la compuerta AND
# (solo 1 cuando ambos son 1)
y = [1,0,0,0]


# PARÁMETROS INICIALES DEL PERCEPTRÓN

pesos = [2, 3]             # Pesos iniciales (pueden ser aleatorios)
sesgo = 4                  # Sesgo inicial
tasa_de_aprendizaje = 0.1  # Cuánto se ajustan los pesos por error


# CREACIÓN Y ENTRENAMIENTO DEL PERCEPTRÓN

# Se crea el objeto 'p' del tipo Perceptron
p = Perpceptron(pesos, sesgo, tasa_de_aprendizaje)

# Se entrena el modelo con los datos de entrada x y salidas y
# durante 20 iteraciones (épocas)
p.entrenamiento(x, y, 20)
```

# ***PERCEPTRÓN OR***

```python
# Se crea la clase perpceotron
class Perpceptron:
    def __init__(self, pesos, sesgo, tasa_de_aprendizaje):
        self.pesos = pesos
        self.sesgo = sesgo
        self.tasa_de_aprendizaje = tasa_de_aprendizaje

    # Función que decide si la salida es 1 o 0
    def funcion_de_activacion(self, z):
        if z >= 0:
            return 1
        else:
            return 0

    # Proceso de aprendizaje
    def entrenamiento(self, x, y, epocas):
        for epoca in range(epocas):
            print(f"\n--- Época {epoca + 1} ---")
            for i, dato in enumerate(x):
                x1, x2 = dato  # Entradas
                w1, w2 = self.pesos
                b = self.sesgo

                #  suma
                z = x1 * w1 + x2 * w2 + b

                #  salida del perceptrón
                y_pred = self.funcion_de_activacion(z)


                y_real = y[i]

                #  ajuste de pesos y sesgo
                self.pesos[0] = w1 + self.tasa_de_aprendizaje * (y_real - y_pred) * x1
                self.pesos[1] = w2 + self.tasa_de_aprendizaje * (y_real - y_pred) * x2
                self.sesgo = b + self.tasa_de_aprendizaje * (y_real - y_pred)

                print(f"Dato: {dato}, Predicción: {y_pred}, Pesos: {self.pesos}, Sesgo: {self.sesgo}")



# OR = 1 cuando al menos uno de los dos es 1
x = [[1,1], [1,0], [0,1], [0,0]]
y = [1,1,1,0]

# Pesos y sesgo iniciales
pesos = [0.5, 0.5]
sesgo = -0.2
tasa_de_aprendizaje = 0.1

# Crear y entrenar el modelo OR
p_or = Perpceptron(pesos, sesgo, tasa_de_aprendizaje)
p_or.entrenamiento(x, y, 10)
```
