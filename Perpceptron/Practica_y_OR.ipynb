# ***PRACTICA DE CLASE***



```python

# Se crea la clase Perceptron
class Perceptron:

    # Esta parte se ejecuta al crear el perceptrón
    def __init__(self, pesos, sesgo, tasa_de_aprendizaje):
        # Guardamos los valores iniciales
        self.pesos = pesos                  # Pesos (importancia de cada entrada)
        self.sesgo = sesgo                  # Sesgo (ajuste adicional)
        self.tasa_de_aprendizaje = tasa_de_aprendizaje  # Qué tan rápido aprende

    # Esta función decide si la salida es 0 o 1
    def funcion_de_activacion(self, z):
        # Si el resultado es positivo o cero → 1
        if z >= 0:
            return 1
        # Si es negativo → 0
        else:
            return 0

    # Aquí el perceptrón aprende
    def entrenamiento(self, x, y, epocas):
        # Se repite el proceso varias veces o sea las épocas
        for epoca in range(epocas):
            print(f"\n--- Época {epoca + 1} ---")


            for i, dato in enumerate(x):
                x1, x2 = dato        # Entradas (dos valores)
                w1, w2 = self.pesos  # Pesos actuales
                b = self.sesgo       # Sesgo actual

                # Se multiplica entradas por pesos y sumamos el sesgo
                z = x1 * w1 + x2 * w2 + b

                # Aplicamos la función de activación
                y_pred = self.funcion_de_activacion(z)

                # Salida
                y_real = y[i]

                # Actualizamos los pesos y el sesgo

                self.pesos[0] = w1 + self.tasa_de_aprendizaje * (y_real - y_pred) * x1
                self.pesos[1] = w2 + self.tasa_de_aprendizaje * (y_real - y_pred) * x2
                self.sesgo = b + self.tasa_de_aprendizaje * (y_real - y_pred)

                # Mostramos los resultados
                print(f"Dato: {dato}, Predicción: {y_pred}, Pesos: {self.pesos}, Sesgo: {self.sesgo}")


# Entradas posibles (0 o 1)
x = [[1,1], [1,0], [0,1], [0,0]]

# Salidas esperadas (solo da 1 cuando los dos son 1)
y = [1,0,0,0]

# Valores iniciales
pesos = [2, 3]              # Pesos de las entradas
sesgo = 4                   # Valor de inicio del sesgo
tasa_de_aprendizaje = 0.1   # Qué tanto se corrigen los errores

# Creamos el perceptrón con esos valores
p = Perceptron(pesos, sesgo, tasa_de_aprendizaje)

# Entrenamos el perceptrón
p.entrenamiento(x, y, 20)
```

# ***PERPCEPTRON DE OR***

```python
# Se crea la clase perpceotron
class Perpceptron:
    def __init__(self, pesos, sesgo, tasa_de_aprendizaje):
        self.pesos = pesos
        self.sesgo = sesgo
        self.tasa_de_aprendizaje = tasa_de_aprendizaje

    # Función que decide si la salida es 1 o 0
    def funcion_de_activacion(self, z):
        if z >= 0:
            return 1
        else:
            return 0

    # Proceso de aprendizaje
    def entrenamiento(self, x, y, epocas):
        for epoca in range(epocas):
            print(f"\n--- Época {epoca + 1} ---")
            for i, dato in enumerate(x):
                x1, x2 = dato  # Entradas
                w1, w2 = self.pesos
                b = self.sesgo

                #  suma
                z = x1 * w1 + x2 * w2 + b

                #  salida del perceptrón
                y_pred = self.funcion_de_activacion(z)


                y_real = y[i]

                #  ajuste de pesos y sesgo
                self.pesos[0] = w1 + self.tasa_de_aprendizaje * (y_real - y_pred) * x1
                self.pesos[1] = w2 + self.tasa_de_aprendizaje * (y_real - y_pred) * x2
                self.sesgo = b + self.tasa_de_aprendizaje * (y_real - y_pred)

                print(f"Dato: {dato}, Predicción: {y_pred}, Pesos: {self.pesos}, Sesgo: {self.sesgo}")



# OR = 1 cuando al menos uno de los dos es 1
x = [[1,1], [1,0], [0,1], [0,0]]
y = [1,1,1,0]

# Pesos y sesgo iniciales
pesos = [0.5, 0.5]
sesgo = -0.2
tasa_de_aprendizaje = 0.1

# Crear y entrenar el modelo OR
p_or = Perpceptron(pesos, sesgo, tasa_de_aprendizaje)
p_or.entrenamiento(x, y, 10)
```
