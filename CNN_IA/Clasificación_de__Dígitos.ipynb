```python
"""
El código muestra una red convolucional
"""

from sklearn.datasets import load_digits

# Cargo el dataset de dígitos
digits = load_digits()


digits.keys()

dict_keys([
    'data', 'target', 'frame',
    'feature_names', 'target_names', 'images',
    'DESCR'
])

# Descripción general del dataset
print(digits.DESCR)

# Tomo un ejemplo para visualizarlo
index = 0
image = digits.images[index]
label = digits.target[index]

print(image)

import matplotlib.pyplot as plt

# Muestro la imagen en escala de grises
plt.figure(figsize=(3, 3))
plt.imshow(image, cmap=plt.cm.gray)
plt.title(f'Dígito: {label}')
plt.show()

import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

# Divido los datos en entrenamiento y prueba (80% - 20%)
X_train, X_test, y_train, y_test = train_test_split(
    digits.data, digits.target,
    test_size=0.2, random_state=42
)

# Convierto las etiquetas a formato
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

print(X_train.shape)
print(X_test.shape)


reshaped_tensor = tf.reshape(X_train[0], shape=(8, 8))
print(reshaped_tensor)

from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Normalizo los datos para que el modelo aprenda mejor
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# normalizar
reshaped_tensor = tf.reshape(X_train[0], shape=(8, 8))
reshaped_tensor = tf.floor(reshaped_tensor * 100) / 100
print(reshaped_tensor)

# Doy forma a los datos como imágenes
X_train = tf.reshape(X_train, (X_train.shape[0], 8, 8))
X_test = tf.reshape(X_test, (X_test.shape[0], 8, 8))

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D

# Creo la de red convolucional
model = Sequential([
    Conv2D(32, kernel_size=(3, 3),
])

from tensorflow.keras.optimizers import Adam

# Defino el optimizador Adam con tasa de aprendizaje
learning_rate = 0.001
adam_optimizer = Adam(learning_rate=learning_rate)

# Compilo el modelo con pérdida y métrica de exactitud
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Entreno el modelo con los datos
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)

# Grafico cómo fue la pérdida en cada época
plt.plot(history.history['loss'], label='Pérdida de entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida de validación')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.legend()
plt.title('Función de pérdida durante el entrenamiento')
plt.show()

# Evalúo el modelo con los datos de prueba
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Loss: {loss}, Accuracy: {accuracy}')

import numpy as np
from sklearn.metrics import confusion_matrix, recall_score

# Hago predicciones con el modelo
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_test_classes = np.argmax(y_test, axis=1)

# Calculo matriz de confusión y sensibilidad (recall)
conf_matrix = confusion_matrix(y_test_classes, y_pred_classes)
sensitivity = recall_score(y_test_classes, y_pred_classes, average=None)

import seaborn as sns
import matplotlib.pyplot as plt

# Muestro la matriz de confusión en colores
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

# Imprimo el recall
print('Sensitivity (Recall) for each class:')
for i in range(10):
    print(f'Class {i}: {sensitivity[i]}')

from PIL import Image
import matplotlib.pyplot as plt

# Cargo una imagen externa para probar la red
ruta = "/content/mi_numero.png"
img = Image.open(ruta).convert("L")  # Escala de grises

# Muestro la imagen original
plt.figure(figsize=(4, 4))
plt.imshow(img, cmap="gray")
plt.title("Imagen original")
plt.axis("off")
plt.show()

import matplotlib.pyplot as plt


img_resized = img.resize((8, 8), Image.Resampling.LANCZOS)

plt.figure(figsize=(3, 3))
plt.imshow(img_resized, cmap="gray")
plt.title("Imagen redimensionada a 8x8")
plt.axis("off")
plt.show()

# Guardo la imagen
img_resized.save("imagen_8x8.png")

import matplotlib.pyplot as plt
import numpy as np


img_array = np.array(img_resized).astype("float32")
img_array = 16 - (img_array / 255 * 16)

# Muestro la imagen procesada
plt.imshow(img_array, cmap="gray")
plt.title("Imagen reescalada e invertida (0-16)")
plt.axis("off")
plt.show()

from sklearn.preprocessing import StandardScaler

# Normalizo la imagen y la preparo para predecir
scaler = StandardScaler()
img_flat = img_array
img_scaled = scaler.fit_transform(img_flat)

# Cambio su forma
img_tensor = img_scaled.reshape(1, 8, 8)

# Hago la predicción final del número
pred = model.predict(img_tensor)
digit = np.argmax(pred)
print("Predicción:", digit)
```
