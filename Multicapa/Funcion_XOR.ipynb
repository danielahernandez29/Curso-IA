```python
"""
Red Neuronal Multicapa para la función XOR usando TensorFlow y Keras
"""

import numpy as np # Necesario para arrays.

# Datos XOR (mis entradas X y salidas y deseadas)
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Definimos nuestra red neuronal multicapa usando Keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

modelo = Sequential([
    # Capa oculta: 4 neuronas y ReLU. Clave para separar los datos.
    Dense(4, activation='relu', input_dim=2),
    # Capa de salida: 1 neurona y Sigmoid (para clasificación binaria 0 o 1).
    Dense(1, activation='sigmoid')
])

# Creamos el optimizador Adam con una tasa de aprendizaje personalizada
from tensorflow.keras.optimizers import Adam

# Tasa de aprendizaje que elegí
learning_rate = 0.01
adam_optimizer = Adam(learning_rate=learning_rate)

# Compilamos
modelo.compile(
    optimizer=adam_optimizer,
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Entrenamos el modelo
# Solo 50 épocas
history = modelo.fit(
    X, y,
    epochs=50,
    verbose=0,
)

# Graficamos la función de pérdida
import matplotlib.pyplot as plt

# Grafico la pérdida (Loss) durante el entrenamiento
plt.plot(history.history['loss'])
plt.title('Evolución de la pérdida (Loss)')
plt.xlabel('Época')
plt.ylabel('Binary Cross-Entropy')
plt.show()

# Evaluamos el modelo en los datos de entrenamiento
loss, accuracy = modelo.evaluate(X, y, verbose=0)
print(f'Loss: {loss}, Accuracy: {accuracy}')

# Hacemos predicciones finales
# Convierto la probabilidad (ej. 0.9) a 1 si es mayor a 0.5
pred = (modelo.predict(X) > 0.5).astype(int)
# Muestro las entradas
for i, (inp, p) in enumerate(zip(X, pred)):
    print(f"{inp} -> {p[0]}")
```
